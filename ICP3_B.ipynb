In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
car_sales = pd.read_csv('car-sales (1).csv')

# Step 1: Clean the 'Price' column
# Remove '$' and ',' symbols and remove last 3 characters (".00"), then convert to integer
car_sales['Price'] = car_sales['Price'].str.replace('[\$,]', '', regex=True).str[:-3].astype(int)

# Print cleaned 'Price' column
print("Cleaned 'Price' column:")
print(car_sales['Price'])
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)

# View the car_sales DataFrame
import pandas as pd

# Step 1: Load the CSV file
car_sales = pd.read_csv('car-sales (1).csv')

# Step 2: Clean the 'Price' column
# - Remove $ and commas
# - Remove the last 3 characters (".00")
# - Convert to integer
car_sales['Price'] = car_sales['Price'].str.replace('[\$,]', '', regex=True).str[:-3].astype(int)

# Step 3: Add 'Total Sales' column using cumulative sum of 'Price'
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Step 4: Add 'Sale Date' column with one date per row starting from today
car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales))

# Step 5: View the updated DataFrame
print(car_sales)
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
car_sales = pd.read_csv('car-sales (1).csv')

# Step 1: Clean the 'Price' column
car_sales['Price'] = car_sales['Price'].str.replace('[\$,]', '', regex=True).str[:-3].astype(int)

# Step 2: Add 'Total Sales' column
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Step 3: Add 'Sale Date' column
car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales))

# Step 4: Plot 'Sale Date' vs 'Total Sales'
car_sales.plot(x='Sale Date', y='Total Sales', marker='o', title='Total Car Sales Over Time')

# Optional: Customize labels
plt.xlabel('Date of Sale')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like plot3

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot


# Plot the data


# Customize the plot


# Add a meanline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

df_over50 = heart_disease[heart_disease['age'] > 50]

plt.figure(figsize=(10, 6))

scatter = plt.scatter(
    df_over50['age'],
    df_over50['chol'],   # changed here
    c=df_over50['target'],
    cmap='winter',
    alpha=0.7,
    edgecolor='k'
)

plt.colorbar(scatter, label='Target')

plt.title('Age vs Cholesterol (chol) Colored by Target (Patients > 50)')
plt.xlabel('Age')
plt.ylabel('Cholesterol')

mean_age = df_over50['age'].mean()
mean_chol = df_over50['chol'].mean()   # changed here

plt.axvline(mean_age, color='green', linestyle='--', label=f'Mean Age: {mean_age:.1f}')
plt.axhline(mean_chol, color='red', linestyle='--', label=f'Mean Cholesterol: {mean_chol:.1f}')

plt.legend()
plt.show()
```
