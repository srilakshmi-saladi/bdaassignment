### Task 17

Alice, Mark, Bill, Tony, Jason, Ravi, Sandeep, Mounika, John and Curtis took three graduate courses `CS 5000`, `CS 5110` and `CS 5600` in Spring 2022. To simulate the midterm scores they got for the courses, please use a random seed 123 to generate a $10 \times 3$ DataFrame called scores to store the 10 stduents' scores. The scores are random integers in the range of `[30, 100]`

```python
np.random.seed(123)

students = ['Alice', 'Mark', 'Bill', 'Tony', 'Jason', 'Ravi', 'Sandeep', 'Mounika', 'John', 'Curtis']
courses = ['CS 5000', 'CS 5110', 'CS 5600']

scores_data = np.random.randint(30, 101, size=(10, 3))

scores = pd.DataFrame(scores_data, index=students, columns=courses)

scores
```

### Task 18

Repeat Task 17 to generate the final scores for those stduents. Please use a random seed of 456 to generate the data

```python
np.random.seed(456)

final_scores_data = np.random.randint(30, 101, size=(10, 3))

final_scores = pd.DataFrame(final_scores_data, index=students, columns=courses)

final_scores
```

### Task 19
Concatenate the two DataFrames from Task 17 and Task 18 into a new DataFrame with multi index

```python

all_scores = pd.concat([scores, final_scores],
                       keys=['Midterm', 'Final'])
all_scores
```

### Task 20
Name the multi index with `Term` and `Name`, respectively

```python
all_scores = pd.concat([scores, final_scores],
                       keys=['Midterm', 'Final'],
                       names=['Term', 'Name'])
all_scores
```

### Task 21
Swap the level of the multi index

```python
all_scores = all_scores.swaplevel(0, 1).sort_index(level = 0)

all_scores
```

### Task 22
Find the average scores for midterm and final for all three courses

```python

average_scores = all_scores.groupby(level='Term', observed=False).mean()

average_scores
```
